```python
import re
import requests
import tldextract
from lxml import etree
from urlparse import urlparse
```

```python
def get_bepress():
    site = requests.get('http://digitalcommons.bepress.com/subscriber_gallery/')
    elements = etree.HTML(site.text).xpath('//div[@id ="gallery"]/div/ul/li/a')

    return [item.values()[0] for item in elements]
```

```python
stuff = get_bepress()
```

```python
def regex_practice():
    pat = '\/\/\w+\.?\w+-?\w+'
    for link in stuff:
        m = re.search(pat, link)
        try:
            print(m.group(0))
        except AttributeError:
            print('Nope. ' + link)
```

```python
def urlparse_stuff():
    for link in stuff:
        shortname = ''
        x = tldextract.extract(link)
        if x.subdomain and x.subdomain != 'www':
            shortname = shortname + x.subdomain + '_'
        shortname += x.domain
        print('{} is the shortname for {}'.format(shortname, link))
        parts = shortname.split('_')
        class_name = ''
        for part in parts:
            class_name += part.capitalize()
        print(class_name)
```

```python
urlparse_stuff()
```
