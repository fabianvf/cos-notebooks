```python
import requests
from lxml import etree
```

```python
url = 'http://scalar.usc.edu/works/guide/rdf/instancesof/content'
search_url = 'http://scalar.usc.edu/works/guide/rdf/instancesof/content?sq=*'

namespaces = {
    'rdf': 'http://www.w3.org/1999/02/22-rdf-syntax-ns#',
    'scalar': 'http://scalar.usc.edu/2012/01/scalar-ns#',
    'sioc': 'http://rdfs.org/sioc/ns#"',
    'prov': 'http://www.w3.org/ns/prov#',
    'art': 'http://simile.mit.edu/2003/10/ontologies/artstor#',
    'dcterms': 'http://purl.org/dc/terms/',
        'ov': 'http://open.vocab.org/terms/'
}
```

```python
x = requests.get(url).content
```

```python
data = etree.XML(x)
```

```python
records = data.xpath('//rdf:Description', namespaces=namespaces)
```
